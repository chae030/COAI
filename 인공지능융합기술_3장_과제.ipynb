```python
from sklearn import datasets, svm
import numpy as np

d = datasets.load_iris()
s = svm.SVC(gamma=0.1, C=10)
s.fit(d.data, d.target)

idx = np.random.choice(len(d.data), 20, replace=False)
test_x = d.data[idx].copy()
test_y = d.target[idx].copy()

for i in range(20) :
  for k in range(4) :
    test_x[i,k] = test_x[i,k]+np.random.uniform(test_x[i,k]*0.05, test_x[i,k]*0.05)

res = s.predict(test_x)

correct = [i for i in range(len(res)) if res[i] == test_y[i]]
accuacy = len(correct)/len(res)
print("정확률=", accuacy*100, "%")
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt

digit = datasets.load_digits()

for i in range(10) :
  indices = np.where(digit.target == i)[0]
  num_index = np.random.choice(indices)
  plt.figure(figsize=(5, 5))
  plt.imshow(digit.images[i], cmap=plt.cm.gray_r, interpolation='nearest')
  plt.show()
  print(digit.data[num_index])
  print("이 숫자는 ", digit.target[num_index], "입니다.")
```
